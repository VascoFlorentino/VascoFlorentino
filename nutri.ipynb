```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Read the data from a CSV file
data = pd.read_csv('nutri.csv', sep=';', encoding='utf-8')

# Select the columns for visualization
columns = ['Nome SAP', 'Lipidos', 'Saturados', 'Monoinsaturados', 'Polinsaturados', 'Omega 3', 'Ácido Linoleico', 'Ácido Alfa-Linolénico', 'EPA', 'DHA', 'Colesterol', 'Acidos Gordos Trans', 'Alcool', 'Acidos Organicos', 'Hidratos De Carbono', 'Acucares', 'Amido', 'Poliois', 'Fibras', 'Proteinas', 'Sal', 'Vit. A', 'Vit. D', 'Vit. E', 'Vit. C', 'Vit. K', 'B1', 'Riboflavina', 'Niacina', 'B6', 'B9', 'B12', 'Biotina', 'B5', 'Calcio', 'Fosforo', 'Ferro', 'Potássio', 'Magnesio', 'Zinco', 'Iodo', 'Selénio']

# Subset the data with the selected columns
subset_data = data[columns]

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(subset_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Select the columns for visualization
columns = ['Nome SAP', 'Lipidos', 'Saturados', 'Monoinsaturados', 'Polinsaturados', 'Omega 3', 'Ácido Linoleico', 'Ácido Alfa-Linolénico', 'EPA', 'DHA', 'Colesterol', 'Acidos Gordos Trans', 'Alcool', 'Acidos Organicos', 'Hidratos De Carbono', 'Acucares', 'Amido', 'Poliois', 'Fibras', 'Proteinas', 'Sal', 'Vit. A', 'Vit. D', 'Vit. E', 'Vit. C', 'Vit. K', 'B1', 'Riboflavina', 'Niacina', 'B6', 'B9', 'B12', 'Biotina', 'B5', 'Calcio', 'Fosforo', 'Ferro', 'Potássio', 'Magnesio', 'Zinco', 'Iodo', 'Selénio']

# Subset the data with the selected columns
subset_data = data[columns]

# Set the "Nome SAP" column as the index
subset_data.set_index('Nome SAP', inplace=True)

# Plot the bar plot
subset_data.plot(kind='bar', figsize=(12, 8))
plt.xlabel('Nome SAP')
plt.ylabel('Value')
plt.title('Comparison of Nutritional Values')
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.show()
```

```python
# Subset the data with the selected columns
subset_data = data[columns]

# Plot the box plot
plt.figure(figsize=(12, 8))
sns.boxplot(data=subset_data)
plt.xlabel('Nutritional Values')
plt.ylabel('Value')
plt.title('Distribution of Nutritional Values')
plt.xticks(rotation=45)
plt.show()
```

```python
# Select the columns for visualization
x_column = 'Lipidos'
y_column = 'Proteinas'

# Plot the scatter plot
plt.figure(figsize=(8, 8))
sns.scatterplot(data=data, x=x_column, y=y_column)
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.title('Scatter Plot: Lipidos vs. Proteinas')
plt.show()
```
